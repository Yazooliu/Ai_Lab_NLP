```python
".data/in_the_name_of_people.txt"
"1.先收集数据, 2: 除去非中文部分，3：处理中文编码问题，4：中文分词"
"中文的编码格式不是utf-8,而是unicode,但是python不支持unicode的处理。所以读取文件时候用GBK解码，存取文件时用utf-8编码"
# -*- coding:utf-8 
import jieba 
with open('./data/nlp_text0.txt') as f1:
    document        = f1.read()
    document_decode = document.decode('GBK')
    
    document_cut    = jieba.cut(document_decode)
    result          = " ".join(document_cut)
    
    result_encode   = result.encode('utf-8')
    with open('./data/nlp_text1.txt','w') as f2:
        f2.write(result_encode)
f2.close()
f1.close()
```

```python
"载入停用词"
import pandas as pd
stpwrd_path = "./data/stop_words_datapreprocessing_1208.txt"
stpwrd    = open("./data/stop_words_datapreprocessing_1208.txt",'rb')
stpwrd_df = pd.read_csv("./data/stopwords_NLP.txt",
                             index_col = False,quoting = 3,sep = '\t',encoding = 'utf-8')
stpwrd_read  = stpwrd.read()

"stopwords转成list"
stpwrd_list  =  stpwrd_read.splitlines()  # stpwrd_read is str,需要使用splitlines()
stpwrd.close()
```

```python
"特征处理 - 由两种特征处理方法：向量化and Hash Trick"
"这里使用scikit learning 中的TfidfVectorizer"
#print "TfidfVectorizer类可以帮助我们完成向量化，TF-IDF和标准化三步。当然，还可以帮我们处理停用词"

"特征处理过程"
with open('./data/nlp_text1.txt') as f3:
    nlp_text1 = f3.read()
#print nlp_text1

with open('./data/nlp_text3.txt') as f4:
    nlp_text3 = f4.read()
#print nlp_text3

from sklearn.feature_extraction.text import TfidfVectorizer 
corpus  = [nlp_text1,nlp_text3]
vector  = TfidfVectorizer(stop_words  = stpwrd_list)
tfidf   = vector.fit_transform(corpus)

wordlist = vector.get_feature_names() # 获取词袋中的所有词

"tf-idf 矩阵 元素a[i][j] 表示j词在i类文本中的tf-idf权重"
weightlist = tfidf.toarray()

"便利每类文本的if - idf词语权重， "
for i in range(len(weightlist)): # 遍历所有文本
    print "------------第",i,"段文本的词语tf-idf权重-----"
    
    for j in range(len(wordlist)):
        print wordlist[j],weightlist[i][j]
```

```python
"英文文本 处理过程："
"英文文本不必像中文那样做分词"
"英文文本都是utf-8编码，不必做编码的转换"
"1.数据收集 2.去除文本中非文本部分 3.拼写检查 4.词干提取和词性还原 5.转换为小写 6.引入停用词 7.特征处理 8.建立分析模型"
import nltk
nltk.download()
```

```python
"直接读入文本数据，不必做分词与编码处理"
with open('./data/nlp_englishtext0.txt') as f1:
    document1 = f1.read()
    
f1.close()
with open('./data/nlp_englishtext0.txt') as f2:
    document2 = f2.read()
    
f2.close()

"输入的整体文本数据"
document_corpos = [document1,document2]

"载入英文stopwords"
import pandas as pd
stpwrds_english = pd.read_csv("./data/uwm-workshop/data/stoplist_english.csv",names = ['stopwords']) 
stpwrds_content = stpwrds_english.stopwords.values.tolist()
```

```python
"通过TfidfVectotizer 来做完成向量化，TF-IDF，标准化三步"
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf = TfidfVectorizer(stop_words = stpwrds_content)

tfidf = Tfidf.fit_transform(document_corpos)  # 权重
#print tfidf

"获取每个词的Tf-IDF 关系"
wordlist = Tfidf.get_feature_names()
weightlist = tfidf.toarray()
for j in range(len(weightlist)):
    print "第",j,"段文本的单词的Tf-IDF 权重是:"
    for i in range(len(wordlist)):
        print wordlist[i],weightlist[j][i]
```

### 词干提取(stemming) 和 词型还原(lemmatization) 是英文文本预处理的特点

#### （1）在nltk中对英文的词干提取有 PorterStemmer/LancasterStemmer/SnowballStemmer, 个人推荐用SnowballStemmer

```python
"SnowballStemmer 提取词干，但是提取的词干并 不是一个词。例如："
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")  # 选择语言类型
stemmer.stem("imaging")
```
